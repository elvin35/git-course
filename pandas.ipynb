```python
students_performance = students_performance.rename(columns= 
                                             {'parental level of education':'parental_level_of_education',
                                               'test preparation course':'test_preparation_course',
                                               'math score':'math_score','reading score':'reading_score',
                                             'writing score':'writing_score'})
```

```python
writing_score_query = 90
```

```python
students_performance.query('gender == "female" & writing_score >75')
```

```python
students_performance.query('gender == "male" & writing_score > writing_score_query') #variable nan muqaise elediymize gore islemeyecek
```

```python
students_performance.query('gender == "male" & writing_score > @writing_score_query')
```

```python
[i for i in students_performance.columns if 'score' in i]
```

```python
students_performance.filter(like='score',axis = 1) #alternativ variant
```

```python
students_performance.groupby('gender') \
.aggregate({'writing_score':'mean', 'reading_score': 'mean', 'math_score':'mean'})
```

```python
students_performance.pivot_table(values=  ['writing_score','reading_score','math_score'], index = 'gender',
                                aggfunc= [np.median,np.mean]) # groupby'n alternativ yolu
```

```python
students_performance.groupby('gender',as_index=False) \
.aggregate({'writing_score':'mean', 'reading_score': 'mean', 'math_score':'mean'}) \
.rename(columns={'writing_score':'mean_writing_score', 
               'reading_score':'mean_reading_score' ,
               'math_score': 'mean_math_score' })
```

```python
mean_scores = students_performance.groupby(['gender', 'race/ethnicity']) \
    .aggregate({'math_score': 'mean', 'reading_score': 'mean'}) \
    .rename(columns = {'math_score': 'mean_math_score', 'reading_score': 'mean_reading_score'})
```
